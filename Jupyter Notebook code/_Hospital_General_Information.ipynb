```python
df['Address'] = df['Address'].str.replace('2505 U S HIGHWAY 431 NORTH','2505 US HIGHWAY 431 NORTH')
```

```python
# Convert to uppercase and remove extra spaces
df['Address'] = df['Address'].str.upper().str.strip().str.replace(r'\s+', ' ', regex=True)
df
```

```python
# Clean the address first
df['clean_address'] = df['address'].astype(str).str.upper().str.strip().str.replace(r'\s+', ' ', regex=True)

# Extract street number
df['street_number'] = df['clean_address'].str.extract(r'^(\w+)')

# Extract direction at the end of address
df['direction'] = df['clean_address'].str.extract(r'(NORTH|SOUTH|EAST|WEST|NE|NW|SE|SW)$')

# Extract street type (optional, could be missing)
df['street_type'] = df['clean_address'].str.extract(r'\b(AVENUE|STREET|ST|DRIVE|DR|RD|ROAD|LANE|LN|HIGHWAY|HWY)\b')

# Initialize street_name as clean_address
df['street_name'] = df['clean_address']

# Remove street_number from street_name (only if not NaN)
df.loc[df['street_number'].notna(), 'street_name'] = df.loc[df['street_number'].notna(), 'street_name'].str.replace(rf'^{df["street_number"]}', '', regex=True)

# Remove direction if present
df.loc[df['direction'].notna(), 'street_name'] = df.loc[df['direction'].notna(), 'street_name'].str.replace(rf'\b{df["direction"]}\b$', '', regex=True)

# Remove street type if present
df.loc[df['street_type'].notna(), 'street_name'] = df.loc[df['street_type'].notna(), 'street_name'].str.replace(rf'\b{df["street_type"]}\b', '', regex=True)

# Final cleanup
df['street_name'] = df['street_name'].astype(str).str.strip()


df
```

```python
import re

def clean_hospital_name(name):
    name = name.lower().strip()

    # Normalize common terms
    name = name.replace("'", "")  # remove apostrophes
    name = name.replace("hospitals", "hospital")  # plural to singular
    name = name.replace("medical center", "hospital")  # standardize term

    # Normalize "St." and "St"
    name = re.sub(r'\bst\.?\s*', 'st ', name)  # Normalize St., St to "st"

    # Specific replacements to unify names
    replacements = {
        "memorial medical hospital": "memorial hospital",
        "memorial medical center": "memorial hospital",
        "community memorial hospital": "community hospital",
        "mercy medical center": "mercy hospital",
        "mercy regional medical center": "mercy hospital",
        "shriners hospital for children": "shriners hospital",
        "shriners hospitals for children": "shriners hospital",
        "st marys hospital": "st mary hospital",
        "st mary's hospital": "st mary hospital",
        "st mary medical center": "st mary hospital",
        "st joseph medical center": "st joseph hospital",
        "st josephs hospital": "st joseph hospital",
        "st lukes hospital": "st luke hospital",
        "northwest medical center": "northwest hospital"
    }

    for wrong, right in replacements.items():
        name = name.replace(wrong, right)

    name = re.sub(r'\s+', ' ', name)  # remove any extra spaces
    return name.strip()

    
df['cleaned_hospital_name'] = df['hospital_name'].apply(clean_hospital_name)
```
